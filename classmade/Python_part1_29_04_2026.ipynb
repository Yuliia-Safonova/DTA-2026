```python
# Калькулятор середнього балу. Спитати ім'я, 3 оцінки, порахувати середній бал, вивести і привітати, якщо склав

name = input("What is your name? ")

grade1 = int(input("What is your first grade? "))
grade2 = int(input("What is your second grade? "))
grade3 = int(input("What is your third grade? "))

average = (grade1 + grade2 + grade3)/3

print("Hello,", name, ". Nice to meet you!")
print("Your average grade is: ", average)

if average >= 60:
  print("Congratulations, you passed!")
else:
  print("Sorry, you need to retake the exam.")
```
