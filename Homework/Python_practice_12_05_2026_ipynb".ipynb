# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

Дано словник з товарами магазину, де значення — це інший словник з ціною і кількістю на складі. Виконайте:
1. Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).
2. Виведіть назви товарів, яких залишилось менше 20 штук.
3. Виведіть найдорожчий товар.

````python
inventory = {
    "Ноутбук":   {"price": 18500, "stock": 12},
    "Телефон":   {"price": 9200,  "stock": 35},
    "Планшет":   {"price": 7800,  "stock": 8},
    "Навушники": {"price": 850,   "stock": 120},
    "Клавіатура":{"price": 1200,  "stock": 45},
    "Мишка":     {"price": 450,   "stock": 18}
}
# Ваш код:

total_suma = 0
low_stock = []
max_name = ""
max_price = 0

for name, info in inventory.items():

    price = info["price"]
    stock = info["stock"]
    total_suma += price * stock

    if stock < 20:
        low_stock.append(name)

    if price > max_price:
        max_price = price
        max_name = name

print(f"1. Загальна вартість всього товару на складі: {total_suma} грн.")
print(f"2. Наступних товарів залишилось < 20 на складі: {low_stock}")
print(f"3. Найдорожчий товар на складі: {max_name} - {max_price} грн.")
````

### Завдання 22. Телефонна книга з пошуком ⭐⭐⭐

Створіть словник-телефонну книгу з 5 контактами. Напишіть програму, яка у циклі `while`:
1. Запитує у користувача дію: `пошук`, `додати`, `видалити`, `вихід`
2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
# Ваш код для завдання 22:
phonebook = {
    "Олена": "+380501234567",
    "Петро": "+380671234567",
    "Ірина": "+380931234567",
    "Дмитро": "+380961234567",
    "Тетяна": "+380731234567",
}

while True:
    action = input("Напишіть, що потрібно зробити (пошук / додати / видалити / вихід): ").strip().lower()

    if action == "пошук":
        name = input("Введіть ім'я: ").strip()
        if name in phonebook:
            print(f"{name}: {phonebook[name]}")
        else:
            print(f"Контакт '{name}' не знайдено.")

    elif action == "додати":
        name  = input("Ім'я: ").strip()
        phone = input("Телефон: ").strip()
        phonebook[name] = phone
        print(f"Контакт '{name}' додано.")

    elif action == "видалити":
        name = input("Введіть ім'я: ").strip()
        if name in phonebook:
            del phonebook[name]
            print(f"Контакт '{name}' видалено.")
        else:
            print(f"Контакт '{name}' не знайдено.")

    elif action == "вихід":
        print("До побачення!")
        break

    else:
        print("Невідома дія. Спробуйте ще раз.")
````

````python
# Ваш код для завдання 26:
history = []

while True:
    operation = input("Введіть операцію ( +, -, *, / ) або 'вихід': ")

    if operation == "вихід":
        break

    num1 = float(input("Введіть перше число: "))
    num2 = float(input("Введіть друге число: "))

    if operation == "+":
        result = num1 + num2

    elif operation == "-":
        result = num1 - num2

    elif operation == "*":
        result = num1 * num2

    elif operation == "/":
        if num2 != 0:
            result = num1 / num2
        else:
            result = "Помилка: ділення на нуль"

    else:
        result = "Невідома операція"

    finish = f"{num1} {operation} {num2} = {result}"

    history.append(finish)

print("\nІсторія Ваших обчислень:")

for el in history:
    print(el)
````

Дано список матчів. Кожен матч — це словник `{"team_a": ..., "team_b": ..., "score_a": ..., "score_b": ...}`. Створіть таблицю очок: за перемогу — 3 очка, за нічию — 1, за поразку — 0. Виведіть команди у порядку від найбільшої до найменшої кількості очок.

> 💡 **Підказка:** Створіть словник `points = {}`. У циклі по матчах: визначайте переможця і додавайте очки. Не забудьте обробити випадок, коли команди ще немає у словнику.

````python
matches = [
    {"team_a": "Динамо",     "team_b": "Шахтар",     "score_a": 1, "score_b": 2},
    {"team_a": "Зоря",       "team_b": "Динамо",     "score_a": 0, "score_b": 1},
    {"team_a": "Шахтар",     "team_b": "Дніпро",     "score_a": 3, "score_b": 1},
    {"team_a": "Динамо",     "team_b": "Ворскла",    "score_a": 2, "score_b": 2},
    {"team_a": "Шахтар",     "team_b": "Зоря",       "score_a": 4, "score_b": 0},
    {"team_a": "Дніпро",     "team_b": "Ворскла",    "score_a": 1, "score_b": 1},
    {"team_a": "Зоря",       "team_b": "Дніпро",     "score_a": 2, "score_b": 1},
    {"team_a": "Ворскла",    "team_b": "Шахтар",     "score_a": 0, "score_b": 3}
]
# Ваш код:

points = {}

for match in matches:
    team_a = match["team_a"]
    team_b = match["team_b"]
    score_a = match["score_a"]
    score_b = match["score_b"]

    if team_a not in points:
        points[team_a] = 0
    if team_b not in points:
        points[team_b] = 0
    if score_a > score_b:
        points[team_a] += 3
    elif score_b > score_a:
        points[team_b] += 3
    else:
        points[team_a] += 1
        points[team_b] += 1

sorted_points = sorted(points.items(), key=lambda x: x[1], reverse=True)

for team, pts in sorted_points:
    print(team, "-", pts)
````

````python
# Ваш код для завдання 44:
alphabet = "абвгґдеєжзиіїйклмнопрстуфхцчшщьюя"
cipher = {}

# Створення словника шифру
for i in range(len(alphabet)):
    current_letter = alphabet[i]
    next_letter = alphabet[(i + 1) % len(alphabet)]
    cipher[current_letter] = next_letter

word = input("Введіть слово: ").lower()
encrypted_word = ""

for letter in word:
    if letter in cipher:
        encrypted_word += cipher[letter]
    else:
        encrypted_word += letter

print("Зашифроване слово:", encrypted_word)
````

Створіть змінну `balance = 5000`. У циклі `while` запитуйте у користувача дію:
- `баланс` — показати баланс
- `зняти` — запитати суму і зняти (якщо вистачає коштів)
- `поповнити` — запитати суму і додати до балансу
- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:

balance = 5000
history = []

while True:

    print("\n=== Банкомат ===")
    print("баланс")
    print("зняти")
    print("поповнити")
    print("історія")
    print("вихід")
    action = input("\nОберіть дію: ").lower()

    # Показати баланс
    if action == "баланс":
        print(f"Ваш баланс: {balance} грн")

    # запитати суму і зняти (якщо вистачає коштів)
    elif action == "зняти":
        amount = float(input("Сума для зняття: "))
        if amount <= balance:
            balance -= amount
            history.append({"type": "зняти", "amount": amount})
            print(f"Знято {amount} грн")
        else:
            print("Недостатньо коштів!")

    # запитати суму і додати до балансу
    elif action == "поповнити":
        amount = float(input("Сума поповнення: "))
        balance += amount
        history.append({"type": "поповнити", "amount": amount})
        print(f"Рахунок поповнено на {amount} грн")

    # показати список усіх операцій
    elif action == "історія":
        print("\nІсторія операцій:")
        if len(history) == 0:
            print("Операцій ще немає")
        else:
            for operation in history:
                print(f"{operation['type']} - {operation['amount']} грн")

    # завершити
    elif action == "вихід":
        print("Роботу завершено")
        break

    # Невідома команда
    else:
        print("Невідома команда")
````

### Завдання 46. Менеджер задач (todo) ⭐⭐⭐

Створіть програму-менеджер задач. Задачі зберігаються у списку словників: `{"task": "...", "done": False}`. У циклі `while` запитуйте дію:
- `додати` — додати нову задачу
- `показати` — показати всі задачі з номерами і статусом
- `виконати` — за номером позначити задачу як виконану
- `видалити` — за номером видалити задачу
- `вихід` — завершити

````python
# Ваш код для завдання 46:

tasks = []

while True:

    print("\n--- Менеджер задач ---")
    print("додати")
    print("показати")
    print("виконати")
    print("видалити")
    print("вихід")
    action = input("\nОберіть дію: ").lower()

    # додати нову задачу
    if action == "додати":
        task_name = input("Введіть задачу: ")
        tasks.append({"task": task_name, "done": False})
        print("Задачу додано!")

    # показати всі задачі з номерами і статусом
    elif action == "показати":
        if len(tasks) == 0:
            print("Список задач порожній")
        else:
            print("\nСписок задач:")
            for i, task in enumerate(tasks, start=1):
                if task["done"]:
                    status = "Виконано"
                else:
                    status = "Не виконано"
                print(f"{i}. {task['task']} - {status}")

    # за номером позначити задачу як виконану
    elif action == "виконати":
        number = int(input("Номер задачі: "))
        if 1 <= number <= len(tasks):
            tasks[number - 1]["done"] = True
            print("Задачу виконано!")
        else:
            print("Невірний номер")

    # за номером видалити задачу
    elif action == "видалити":
        number = int(input("Номер задачі: "))
        if 1 <= number <= len(tasks):
            deleted_task = tasks.pop(number - 1)
            print(f"Задачу {deleted_task['task']}' видалено")
        else:
            print("Невірний номер")

    # завершити
    elif action == "вихід":
        print("Програму завершено")
        break

    # Невідома команда
    else:
        print("Невідома команда")
````

### Завдання 52. Гра в слова ⭐⭐⭐

Реалізуйте гру: дано стартове слово. Гравець має ввести наступне слово, яке починається на ОСТАННЮ літеру попереднього. Якщо слово вже використовувалось — програш. Якщо слово починається не на ту літеру — програш. Цикл триває, поки гравець не програє або не введе `здаюсь`.

> 💡 **Підказка:** Зберігайте всі використані слова у списку. Першу і останню літеру отримуйте через індекси: `word[0]` і `word[-1]`.

````python
# Ваш код для завдання 52:

# Стартове слово
current_word = "перемога"

# Список використаних слів
used_words = [current_word]

print("Гра в слова!")
print("Стартове слово: ", current_word, "\n")

while True:
    # Ввід нового слова
    new_word = input("Введіть своє слово або 'здаюсь': ").lower()

    # Завершення гри
    if new_word == "здаюсь":
        print("Гру завершено!")
        break

    # Перевірка повтору
    if new_word in used_words:
        print("Таке слово вже використовувалось!")
        print("Ви програли!")
        break

    # Остання літера попереднього слова
    last_letter = current_word[-1]

    # Перша літера нового слова
    first_letter = new_word[0]

    # Перевірка першої літери
    if first_letter != last_letter:
        print("Слово повинно починатись на літеру", last_letter)
        print("Ви програли!")
        break

    # Додаємо слово у список
    used_words.append(new_word)

    # Нове поточне слово
    current_word = new_word
    print("Слово прийнято!")

# Вивід усіх слів
print("\nВикористані слова:")
for word in used_words:
    print(word)
````
