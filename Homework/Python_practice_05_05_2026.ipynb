1. Чи можна з них скласти трикутник? (правило: сума будь-яких двох сторін має бути більшою за третю)
2. Якщо можна — який це трикутник:
   - рівносторонній (всі сторони рівні)
   - рівнобедрений (дві сторони рівні)
   - різносторонній (всі різні)

> 💡 **Підказка:** Спочатку перевірте, чи трикутник взагалі існує. Потім вже визначайте тип через порівняння сторін.

````python
# Ваш код для завдання 18:
x = float(input("Введіть першу сторону трикутника: "))
y = float(input("Введіть другу сторону трикутника: "))
z = float(input("Введіть третю сторону трикутника: "))
if (x + y > z) and (y + z > x) and (x + z > y):
    print("Вітаю, такий трикутник існує!")
    if x == y == z:
        print("Це трикутник рівносторонній.")
    elif (x == y) or (x == z) or (y == z):
        print("Цей трикутник рівнобедрений")
    else:
        print("Цей трикутник різносторонній (всі сторони різні)")
else:
    print("Вибачте, але не існує трикутника з такими сторонами.")
````

### Завдання 27. Прості числа ⭐⭐⭐

Виведіть всі прості числа від 2 до 50.

**Нагадування:** просте число — це число, яке ділиться тільки на 1 і саме на себе.

> 💡 **Підказка:** Для кожного числа `n` від 2 до 50 перевіряйте, чи має воно дільники від 2 до n-1. Можна використати змінну-прапорець `is_prime = True`.

````python
# Ваш код для завдання 27:
print("Знайдемо всі прості числа від 2 до 50: ")
for n in range(2, 51):
    is_prime = True
    for d in range(2, n):
        if n % d == 0:
            is_prime = False
            break
    if is_prime:
        print(n, end = " ")
````

> 💡 **Підказка:** У циклі: беремо останню цифру через `n % 10`, додаємо до суми, потім відкидаємо її через `n = n // 10`. Цикл працює, поки `n > 0`.

**Алгоритм:** доки `b ≠ 0`: одночасно `a, b = b, a % b`. Коли `b = 0`, відповідь — це `a`.

**Приклад:** НСД(48, 36) = 12.

````python
# Ваш код для завдання 46:

a = int(input("Введіть перше додатнє число: "))
b = int(input("Введіть друге додатнє число: "))
while b != 0:
    a, b = b, a % b
print("Найбільший спільний дільник цих чисел (НСД) =", a)
````

### Завдання 56. Сортування без `.sort()` ⭐⭐⭐

Дано список чисел. Відсортуйте його за зростанням **без використання методу `.sort()` та функції `sorted()`**.

> 💡 **Підказка:** Найпростіший алгоритм — «бульбашкове сортування»: у вкладеному циклі порівнюємо сусідні елементи і міняємо їх місцями, якщо ліворуч стоїть більший. Повторюємо, поки не пройдемо весь список без жодної заміни.

````python
numbers = [42, 17, 89, 3, 56, 91, 24]
# Ваш код для завдання 56:

n = len(numbers)
for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print("Список, відсортований за зростанням:", numbers)
````

````python
# Ваш код для завдання 59:
print("Гра «Вгадай число»")
secret = 42
guessed = False
t = 0

while not guessed:
    num = int(input("Введіть число від 1 до 100: "))
    t += 1

    if num < secret:
        print("Давай більше! ")
    elif num > secret:
        print("Давай менше! ")
    else:
        guessed = True
        print(f"Вітаю! Ви вгадали це число за {t} спроб(и)!")
````

### Завдання 60. Аналіз тексту ⭐⭐⭐

Дано рядок з реченням. Виконайте такий аналіз і виведіть результати:
1. Кількість літер (без пробілів)
2. Кількість слів
3. Кількість голосних літер
4. Найдовше слово

> 💡 **Підказка:** Метод `.split()` розбиває рядок на список слів. Наприклад, `'hello world'.split()` повертає `['hello', 'world']`.

````python
text = "Програмування на Python це цікаво і корисно"
# Ваш код для завдання 60:

text = input("Напишіть своє речення: ")
print()

letters = 0
for el in text:
    if el != " ":
        letters += 1
print(f"Всього літер в цьому реченні (без пробілів): {letters}\n")

words = text.split()
print(f"Всього слів в цьому реченні: {len(words)}\n")

vowels_uk = "аеєиіїоуюяАЕЄИІЇОУЮЯ"
vowels_en = "aeiouAEIOU"
vowels = 0
for v in text:
    if v in vowels_uk or v in vowels_en:
        vowels += 1
print(f"Голосних літер в цьому реченні: {vowels}\n")

longest = ""
for w in words:
    if len(w) > len(longest):
        longest = w
print(f"Найдовше слово в цьому реченні: «{longest}» - {len(longest)} літер.")
````

### Завдання 62. Турнірна таблиця ⭐⭐⭐

Дано два списки: команди та їхні очки (відповідні позиції).
Виведіть турнірну таблицю, ВІДСОРТОВАНУ за спаданням очок, у такому форматі:
```
1. Шахтар — 67 очок
2. Динамо — 64 очки
...
```

> 💡 **Підказка:** Найпростіший спосіб: у циклі знаходьте команду з максимальними очками, виводьте її, потім видаляйте цю пару (команду і очки) з обох списків. Повторюйте, поки списки не спорожніють.

````python
teams = ["Динамо", "Шахтар", "Зоря", "Дніпро", "Ворскла", "Олександрія"]
points = [64, 67, 52, 48, 39, 45]
# Ваш код для завдання 62:

teams_copy  = teams.copy()
points_copy = points.copy()

print("* ТУРНІРНА ТАБЛИЦЯ ФУТБОЛЬНИХ КОМАНД *\n")

place = 1
while teams_copy:
    best_idx = points_copy.index(max(points_copy))
    best_team   = teams_copy[best_idx]
    best_points = points_copy[best_idx]

    print(f"{place}. {best_team} — {best_points} очок")

    teams_copy.pop(best_idx)
    points_copy.pop(best_idx)
    place += 1
````
