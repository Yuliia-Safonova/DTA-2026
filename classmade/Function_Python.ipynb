````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

NoneType - нам показали результат (або ні), у будь-якому випадку, працювати з результатом повернення функції ми не можемо.

````python
type(None)
````

````python
# Повертає NoneType, без змін даних, виводить результат
def info(d):
    print(d)

# info(my_dict)
type(info(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, виводить результат
def info(d):
    d[6] = "six"
    print(d)

# info(my_dict)
type(info(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, НЕ виводить результат
def add_el(d):
    d[6] = "six"

print(type(add_el(my_dict)))
print(f"my_dict: {my_dict}")
````

````python
print(type(add_el(my_dict)))
display(type(add_el(my_dict)))
````

Функція повертає результат для обробки

````python
def sum_num(a, b):
    return a + b

print(sum_num(5, 7))
print(type(sum_num(5, 7)))

print()

sum_num(5, 7) + 10
````

````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

````python
def dict_value(d, key):
    print(d[key]) # my_dict[3]

dict_value(my_dict, 3)
````

````python
def dict_value(d, key):
    return d[key] # my_dict[3]

print(dict_value(my_dict, 3))
print(type(dict_value(my_dict, 3)))
dict_value(my_dict, 3) + "333"
````

````python
def func_3(a=5, b=7):
    print(a + b)

func_3()        # 5 + 7 = 12
func_3(10)      # 10 + 7 = 17 (a=10)
func_3(10, 20)  # 10 + 20 = 30 (a=10, b=20)
func_3(b=10)    # 5 + 10 = 15 (b=10)
````

````python
def func_4(a, b=7):
    print(a + b)

# func_4()      # TypeError: func_4() missing 1 required positional argument: 'a'
func_4(10)      # 10 + 7 (a=10)
func_4(10, 20)  # 10 + 20
````
