```python
item_price = {
    "Cake": 3.0,
    "Coffee": 2.0,
    "Cookie": 1.0,
    "Juice": 3.0,
    "Salad": 5.0,
    "Sandwich": 4.0,
    "Smoothie": 4.0,
    "Tea": 1.5
}

# price_item = {}
# for key, value in item_price.items():
#   price_item[value] = key

# price_item = {value: key for key, value in item_price.items()}

# price_item = {value: [] for key, value in item_price.items()}
# price_item[3.0].append("Cake")

price_item = {}
for key, value in item_price.items():
  price_item.setdefault(value, []).append(key)

price_item
```

```python
# import random
import numpy as np

RANDOM_STATE = 42
np.random.default_rng(seed=RANDOM_STATE)

price = 3.0
rnd_item = np.random.choice(price_item[price])

print("Випадковий товар:", rnd_item)
print("Ціна обраного товару:", price)
```
