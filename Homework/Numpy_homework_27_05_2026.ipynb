### ✏️ Завдання 1.1

У тебе є дані середньомісячних температур у Харкові за рік (°C):

```
-5, -4, 2, 10, 17, 21, 23, 22, 16, 9, 2, -3
```

1. Створи NumPy масив із цих значень і назви його `temps`
2. Виведи масив
3. Виведи: кількість вимірів (`ndim`), форму (`shape`), кількість елементів (`size`), тип даних (`dtype`)

### ✏️ Завдання 3.1

Є масив цін товарів у доларах. Виконай **без циклів**:

```python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])
```

1. Переведи ціни в гривні (курс: **1 USD = 41.5 UAH**), результат округли до 2 знаків
2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
5. Виведи результати з поясненнями

````python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])

# Твій код:

# Переведи ціни в гривні (курс: 1 USD = 41.5 UAH), результат округли до 2 знаків
prices_uah = np.round(prices_usd * 41.5, 2)
print("Ціни в гривні, UAH:")
print(prices_uah)

# Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
prices_vat = np.round(prices_uah * 1.2, 2)
print("\nЦіни з ПДВ 20%, грн.:")
print(prices_vat)

# Знайди: мінімальну, максимальну, середню ціну з ПДВ
print("\nМінімальна ціна з ПДЗ:", prices_vat.min())
print("Максимальна ціна з ПДЗ:", prices_vat.max())
print("Середня ціна з ПДЗ:", round(prices_vat.mean(), 2))

# Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
print("\nСума всіх товарів з кошика:", prices_vat.sum())
````

### ✏️ Завдання 3.2

Є дані про кількість товарів на складі та їх собівартість.
Виконай **без циклів**:

```python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']
```

1. Порахуй загальну вартість кожного товару на складі (`quantity * cost_each`)
2. Знайди загальну вартість всього складу
3. Знайди товар з найбільшою сумарною вартістю — використай `np.argmax()`
4. Знайди товар з найменшою сумарною вартістю — `np.argmin()`
5. Виведи середню вартість одного найменування на складі

````python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']

# Твій код:

# Порахуй загальну вартість кожного товару на складі (quantity * cost_each)
total_cost = quantity * cost_each
print("Загальна вартість кожного товару:")
print(total_cost)

# Знайди загальну вартість всього складу
warehouse_total = total_cost.sum()
print("\nЗагальна вартість всього складу:", warehouse_total, "грн.")

# Знайди товар з найбільшою сумарною вартістю — використай np.argmax()
max_index = np.argmax(total_cost)
print("\nТовар з найбільшою сумарною вартістю:")
print(names[max_index], "-", total_cost[max_index], "грн.")

# Знайди товар з найменшою сумарною вартістю — np.argmin()
min_index = np.argmin(total_cost)
print("\nТовар з найменьшою сумарною вартістю:")
print(names[min_index], "-", total_cost[min_index], "грн.")

# Виведи середню вартість одного найменування на складі
print("\nСередня вартість одного товару:", round(total_cost.mean(), 2), "грн.")
````

### ✏️ Завдання 4.1

Є масив часу виконання замовлень у хвилинах (норма — до 30 хв):

```python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])
```

1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)
2. Знайди всі **запізнілі** замовлення (> 30 хв)
3. Порахуй кількість вчасних і запізнілих
4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
5. Яка середня затримка серед запізнілих замовлень?

````python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])

# Твій код:

# Знайди всі замовлення, виконані вчасно (≤ 30 хв)
on_time = delivery_time[delivery_time <= 30]
print("Вчасні замовлення:")
print(on_time)

# Знайди всі запізнілі замовлення (> 30 хв)
late = delivery_time[delivery_time > 30]
print("\nЗапізнілі замовлення:")
print(late)

# Порахуй кількість вчасних і запізнілих
print("\nКількість вчасних замовлень:", on_time.size)
print("\nКількість запізнілих замовлень:", late.size)

# Знайди відсоток запізнілих замовлень (округли до 1 знаку)
print("\nВідсоток запізнілих замовлень", round((late.size / delivery_time.size) * 100, 1), "%")

# Яка середня затримка серед запізнілих замовлень?
print("\nСередня затримка:", (late - 30).mean(), "хв.")
````

### ✏️ Завдання 4.2

Є масив щоденних продажів за місяць. Використай `np.where()`:

```python
np.random.seed(42)
daily = np.random.randint(100, 900, size=30)
```

1. Знайди дні з продажами **вище середнього** за місяць
2. Порахуй скільки таких днів
3. За допомогою `np.where()` створи масив `status`:
   - `'Гарний день'` — якщо продажі вище середнього
   - `'Звичайний день'` — якщо нижче або рівно
4. Виведи перші 10 днів: продажі та їх статус поряд

````python
np.random.seed(42)
daily = np.random.randint(100, 900, size=30)

# Твій код:
print("Продажі за місяць:")
print(daily)

# Знайди дні з продажами вище середнього за місяць
avg_sales = daily.mean()
print("\nДні з продажами вище середнього:")
print(daily[daily > avg_sales])

# Порахуй скільки таких днів
print("\nКількість днів вище середнього:", daily[daily > avg_sales].size)

# За допомогою np.where() створи масив status:
    #   'Гарний день' — якщо продажі вище середнього
    #   'Звичайний день' — якщо нижче або рівно
status = np.where(daily > avg_sales, 'Гарний день', 'Звичайний день')
print("\nСтатуси днів:")
print(status)

# Виведи перші 10 днів: продажі та їх статус поряд
print("\nПерші 10 днів:")

for i in range(10):
    print(f"День {i+1}: продажі = {daily[i]}, статус - {status[i]}")
````
